# Selecting columns

You can select specific columns with `[]`, similarly to how you subset lists and strings, except that instead of integers, you use **column names** (as strings) for selection.

```python
iris["species"]
```

To select multiple columns, pass a `list` of column names.

```python
cols = ["sepal_length", "sepal_width"]
iris[cols]
# iris[["sepal_length", "sepal_width"]]
```

What is the type of a single column?

```python
type(iris["species"])
```

In pandas, a single column is called a **`series`**.

```python
iris[6:8]
```

```python
gt5 = iris[iris["sepal_length"] > 5]
gt5["species"] == "setosa"
```

```python
iris["sepal_area"] = iris["sepal_length"] * iris["sepal_width"]
iris.head()
```

# Split-Apply-Combine

```python
setosa = iris[iris["species"] == "setosa"]
```

```python
iris_grouped = iris.groupby("species")
print iris_grouped
```

```python
iris_grouped.sum()
```

```python
iris_mean = iris_grouped.mean()
iris_sd = iris_grouped.std()
iris_mean
```

```python
iris_grouped.aggregate(["min", "mean", "std", "max"])
```
